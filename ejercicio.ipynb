## Crear una función que devuelva la suma de los n primeros números, en la que el parámetro de entrada es n.

```python
def funcion_1(n):
    """
    Esta función calcula la suma de los números desde 0 hasta n
    """
    resultado = 0
    for i in range(n + 1):
        resultado += i
    return resultado

input_num = int(input("Ingrese un número: "))
suma = funcion_1(input_num)
print(f"La suma de los números desde 0 hasta {input_num} es: {suma}")
```
